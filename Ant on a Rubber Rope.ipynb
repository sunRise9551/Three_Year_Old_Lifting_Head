# 问题

一根弹性均匀的橡皮筋长10米，一只小蚂蚁从一端爬向另一端。小蚂蚁每天沿着橡皮筋爬1米。小蚂蚁每爬1米，橡皮筋就瞬间拉长10米，理想条件下，小蚂蚁不会死，橡皮筋不会被拉断。问：多少天小蚂蚁可以爬到橡皮筋的另一端。

# 思路

提取我们已知的信息：

1. <font color="red">弹性均匀</font>的橡皮筋长10米
2. 小蚂蚁从一端目的爬向另一端
3. 小蚂蚁腼腆可以爬1米
4. 小蚂蚁每爬完1米，橡皮筋拉长10米


## 弹性均匀

这个是一个需要关注的重点。这里<font color="red">均匀</font>的意思是：
- 橡皮筋是<font color="red">不会</font>只在末端/尾部进行拉长
- 橡皮筋<font color="green">会</font>被均匀的拉长（看图）

image.png

## 图理解
- 将 1KM/S 理解为 1米
- 原始长度是1米，第二天拉伸了1米（灰色部分）
- 这个（灰色部分）的1米是会平均分散到整个橡皮筋的长度当中

## 图解释
- 如果我们取0.5来，也就是说每0.5米加长了0.5米 （灰色部分2次）
- 如果我们取0.25米，也就是每0.25米加长了0.25米（灰色部分4次）
- 如果我们取0.1米，也就是每0.1米加长了0.1米   （灰色部分10次）


# 解答

所以只要时间充裕，蚂蚁每天只能完成第一块灰色区域前的那一部分，那么蚂蚁可以爬到橡皮筋的另一端。

# 代码

## 完整代码

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day = 0
ant_percentage = 0
ant_list = []

while ant_percentage <= 1:
  ant_percentage += (1/10) * 1/(day+1)
  day += 1
  ant_list.append(ant_percentage)
```

```python
print("小乌龟在第 {} 天爬到了另一端".format(day))
x = np.linspace(1, day, day)
y = np.array(ant_list)
plt.plot(x, y)
```

## 代码解释

```python
day = 0
# 初始天数 0

ant_percentage = 0
# 我们用比例来表示蚂蚁爬了多少 %，可以理解为蚂蚁爬了伸长后的多少 %

ant_list = []
# 存放每天爬行的 %，为了绘图，没有计算意义

while ant_percentage <= 1:
# 1 表示 100%，只要没有达到 100%， 执行循环到 die ~

  ant_percentage += (1/10) * 1/(day+1)
  # ant_percentage : 蚂蚁爬行的百分比
  # (1/10)    : 蚂蚁每天爬1米，橡皮筋长度10米，两个都是等比例爬行1米和身长10米
  #             (也就是 1/10, 在这个条件下蚂蚁是不可能完成的)
  # 1(day+1)  : 加上均衡拉伸条件，也就是天数越多，每个部分拉伸的区域表少
  #             (可以理解为灰色部分的拉伸长度会随着天数逐渐减少)

  day += 1
  # 每天完成指定 百分比% 的爬行后，计算天数+1

  ant_list.append(ant_percentage)
  # 将每天完成的 百分比% 记录起来，绘图用，没有计算意义
```

```python
# 绘图代码，就不解释了
print("小乌龟在第 {} 天爬到了另一端".format(day))
x = np.linspace(1, day, day)
y = np.array(ant_list)
plt.plot(x, y)
```
